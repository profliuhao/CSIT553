```python
tips = pd.read_csv('tips.csv')
tips['tip_pct'] = tips['tip']/tips['total_bill']
tips
```

```python
grouped = tips['tip_pct'].groupby(tips['day'])
grouped
```

```python
grouped.mean()
```

```python
grouped2 = tips['tip_pct'].groupby([tips['day'], tips['time']])
grouped2.mean()
```

```python
for name, group in tips.groupby('day'):
  print(name)
  print(group)
```

```python
grouped = tips.groupby('smoker')['tip_pct']
grouped.describe()
```
